```python
#bai 13
def tron_mang(a,b):
  kq=[]
  do_dai_ngan_nhat=min(len(a),len(b))
  for i in range(do_dai_ngan_nhat):
    tong=a[i]+b[i]
    kq.append(tong)
  if len(a)>len(b):
    kq.extend(a[do_dai_ngan_nhat:])
  else:
    kq.extend(b[do_dai_ngan_nhat:])
  return kq
mang_a=[3,9,1,4]
mang_b=[2,7, 4, 3, 2, 8]
ketqua=tron_mang(mang_a,mang_b)
print("Ket qua tron hai mang la:",ketqua)
```
